# CLUSTER(K-MEAN)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
data
```

```python
data.isnull().any()
```

```python
X=data.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
                       #weight cluster sample space
```

```python
for i in range(1,20):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,20),wcss)
plt.title("the elbow method")
plt.xlabel("number of cluster")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
```

```python
kmeans
```

```python
ymeans=kmeans.fit_predict(X)
```

```python
ymeans
```

```python
plt.scatter(X[ymeans==0,0],X[ymeans==0,1],s=100,c="red",label="cluster1")
plt.scatter(X[ymeans==1,0],X[ymeans==1,1],s=100,c="blue",label="cluster2")
plt.scatter(X[ymeans==2,0],X[ymeans==2,1],s=100,c="green",label="cluster3")
plt.scatter(X[ymeans==3,0],X[ymeans==3,1],s=100,c="cyan",label="cluster4")
plt.scatter(X[ymeans==4,0],X[ymeans==4,1],s=100,c="magenta",label="cluster5")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c="yellow",label="centroid")
plt.title("cluster of customers")
plt.xlabel("annual income(K$)")
plt.ylabel("spending score (1-100)")
plt.legend()
plt.show()
```
